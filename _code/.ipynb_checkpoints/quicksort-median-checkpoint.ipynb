```python
import itertools

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

import utils

alphas = np.arange(0.1, 0.5, 0.01)
ks = np.arange(1, 21, 2)
```

```python
%matplotlib notebook
x = list(itertools.product(list(ks), list(alphas)))
y = list(map(lambda x: x[1], x))
x = list(map(lambda x: x[0], x))
z = [utils.prob_worse_split(k, alpha) for k, alpha in zip(x, y)]

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_trisurf(x, y, z, cmap=cm.cool);
ax.set_xticks(ks)
ax.set_xlabel(r'$k$')
ax.set_ylabel(r'$\alpha$')
ax.set_zlabel('P[worse split]');
```
